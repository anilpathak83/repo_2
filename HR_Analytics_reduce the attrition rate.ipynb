# Data Visualization

Feature- in df dept and salary are there so we will plot column features against left column.
Columns -satisfactoryLevel-LastEvaluation depend completely on employee and employer.

```python
df1=['dept','numberOfProjects','timeSpent.company','workAccident','promotionInLast5years','salary']
```

```python
df1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig=plt.subplots(figsize=(10,15))
for p,q in enumerate(df1):   # p -index, q values
    plt.subplot(3,2,p+1)
    sns.countplot(x=q, data=df, hue='left')
    plt.subplots_adjust(hspace=1.0)
    plt.xticks(rotation=45)
```

```python
sns.countplot(x=df['dept'])
plt.xticks(rotation=45)
plt.show()
```
